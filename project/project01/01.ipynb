```python
movie_box_week = []
for box_movies in response_list:
    for box_movie in box_movies.values():
         # 1주일치 top 10 영화 정보
        for box_movie_data in box_movie['weeklyBoxOfficeList']:
            movie_box_info = {} # 영화 1개의 정보
            for info, data in box_movie_data.items():
                if info in ['audiAcc', 'movieNm', 'movieCd']:
                    movie_box_info[info] = data
            movie_box_week.append(movie_box_info)
pprint.pprint(movie_box_week)             
```

```python
with open('boxoffice.csv', 'w', encoding='utf-8') as f:
    fieldnames = ['movieCd', 'movieNm', 'audiAcc'] # 여기만 변경하면서 사용?
    csv_writer = csv.DictWriter(f, fieldnames=fieldnames)
    csv_writer.writeheader()
    movie_cd = []
    for item in movie_box_week:
        if item['movieCd'] not in movie_cd:
            csv_writer.writerow(item)
            movie_cd.append(item['movieCd'])
```
